```python
from boto3 import resource, client
```

```python
ec2 = resource("ec2", region_name="us-east-1")
```

```python
# create EC2 instance
ec2.create_instances(ImageId="ami-0fc5d935ebf8bc3bc", InstanceType="t2.micro", MinCount=1, MaxCount=1)
```

```python
# get instances info
ec2.meta.client.describe_instance_status()
```

```python
# get instance complete info
aws = client("ec2", region_name="us-east-1")
aws.describe_instances()
```

```python
# stop instance
ec2.instances.filter(InstanceIds=['i-0fc5d935ebf8bc3bc']).stop()
```

```python
# terminate instance
ec2.instances.filter(InstanceIds=['i-0fc5d935ebf8bc3bc']).terminate()
```

```python
from troposphere import Ref, Template, ec2
```

```python
template = Template()

instancia = ec2.Instance(
    "pythonproserver",
    ImageId="ami-0fc5d935ebf8bc3bc",
    InstanceType="t2.micro"
)

template.add_resource(instancia)
```

```python
from pprint import pprint

pprint(template.to_yaml())
```
